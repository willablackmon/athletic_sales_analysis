```python
# Import Libraries and Dependencies
import pandas as pd
from pathlib import Path
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
csv_path_2020 = Path("Resources/athletic_sales_2020.csv")
csv_path_2021 = Path("Resources/athletic_sales_2021.csv")

sales_2020_df = pd.read_csv(csv_path_2020)
sales_2021_df = pd.read_csv(csv_path_2021)
```

```python
# Display the 2020 sales DataFrame
sales_2020_df.head(10)
```

```python
# Display the 2021 sales DataFrame
sales_2020_df.head(10)
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
sales_2020_df.info()
```

```python
# Check the 2021 sales data types.
sales_2021_df.info()
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.

sales_2020_2021_df = pd.concat([sales_2020_df, sales_2021_df], axis=0)
sales_2020_2021_df                   
```

```python
# Check if any values are null.
sales_2020_2021_df.isnull().sum()
```

```python
# Check the data type of each column
sales_2020_2021_df.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
sales_2020_2021_df['invoice_date'] = pd.to_datetime(sales_2020_2021_df['invoice_date'])
sales_2020_2021_df.head(10)
```

```python
# Confirm that the "invoice_date" data type has been changed.
sales_2020_2021_df.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
sales_region_products_gp = sales_2020_2021_df.groupby(by=['region','state','city'])
sales_region_products_gp_final = sales_region_products_gp[['units_sold']].sum().\
    sort_values('units_sold', ascending=False)

# Rename the sum to "Total_Products_Sold".
sales_region_products_gp_final = sales_region_products_gp_final.rename\
                        (columns={"units_sold":"Total_Products_Sold"})

# Show the top 5 results.
sales_region_products_gp_final.head(5)
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
sales_region_products_pv = pd.pivot_table(sales_2020_2021_df,
                    index=['region','state','city'],
                    values='units_sold',
                    aggfunc='sum')                       

# Rename the "units_sold" column to "Total_Products_Sold"
sales_region_products_pv_final = sales_region_products_pv.rename\
                    (columns={"units_sold":"Total_Products_Sold"})

# Show the top 5 results.
sales_region_products_pv_final.sort_values('Total_Products_Sold',\
                    ascending=False).head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.

sales_region_sales_gp = sales_2020_2021_df.groupby(by=['region','state','city'])
sales_region_sales_gp_final = sales_region_sales_gp[['total_sales']].sum().\
    sort_values('total_sales', ascending=False)

# Rename the "total_sales" column to "Total Sales"
sales_region_sales_gp_final = sales_region_sales_gp_final.\
                            rename(columns={'total_sales':"Total Sales"})
# Show the top 5 results.
sales_region_sales_gp_final.head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
sales_region_sales_pv = pd.pivot_table(sales_2020_2021_df,
                    index=['region','state','city'],
                    values='total_sales',
                    aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
sales_region_sales_pv_final = sales_region_sales_pv.rename\
                    (columns={"total_sales":"Total Sales"})

# Show the top 5 results.
sales_region_sales_pv_final.sort_values('Total Sales',\
                    ascending=False).head(5)
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
retailer_sales_gp = sales_2020_2021_df.groupby(by=\
                ['retailer','region','state','city'])

retailer_sales_gp_final = retailer_sales_gp[['total_sales']].sum().\
                sort_values('total_sales', ascending=False)

# Rename the "total_sales" column to "Total Sales"
retailer_sales_gp_final = retailer_sales_gp_final.rename\
                (columns={"total_sales":"Total Sales"})
                                                  
# Show the top 5 results.
retailer_sales_gp_final.head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
retailer_sales_pv = pd.pivot_table(sales_2020_2021_df,\
                            index=['retailer','region','state','city'],
                            values='total_sales',
                            aggfunc=('sum'))

# Optional: Rename the "total_sales" column to "Total Sales"
retailer_sales_pv_final = retailer_sales_pv.rename\
                    (columns={"total_sales":"Total Sales"})

# Show the top 5 results.
retailer_sales_pv_final.sort_values('Total Sales',\
                    ascending=False).head(5)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
sales_2020_2021_df.head()
```

```python
sales_2020_2021_df.columns
```

```python
# Filter data ales data to get the women's athletic footwear sales data.

# df[df["column_name"].str.contains("<value>")] or 
# df.loc[(df["column_name"] =="<value>")]           (: for all rows)

sales_womens_df = sales_2020_2021_df.loc[\
                    (sales_2020_2021_df['product'] == "Women's Athletic Footwear"), :]

sales_womens_df
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.


units_sold_womens_gb = sales_womens_df.groupby(\
                        by=['retailer','region','state','city'])
units_sold_womens_gb_final = units_sold_womens_gb[['units_sold']].sum().\
                        sort_values('units_sold', ascending=False)

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
units_sold_womens_gb_final = units_sold_womens_gb_final.rename(\
    columns={'units_sold':'Womens_Footwear_Units_Sold'})

# Show the top 5 results.
units_sold_womens_gb_final.head(5)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
units_sold_womens_pv = sales_womens_df.pivot_table(\
                        index=['retailer','region','state','city'], 
                        values='units_sold',
                        aggfunc = 'sum')

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
units_sold_womens_pv_final = units_sold_womens_pv.rename(\
                        columns={'units_sold':'Womens_Footwear_Units_Sold'})

# Show the top 5 results.
units_sold_womens_pv_final = units_sold_womens_pv_final.sort_values(\
                        'Womens_Footwear_Units_Sold',ascending=False).head(5)
units_sold_womens_pv_final
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.

womens_sales_day_df = sales_womens_df.pivot_table(\
                        index=['invoice_date'], 
                        values='total_sales',
                        aggfunc = 'sum')

# Optional: Rename the "total_sales" column to "Total Sales"
womens_sales_day_df_final = womens_sales_day_df.rename(\
                        columns={'total_sales':'Total Sales'})

# Show the table.
womens_sales_day_df_final.head(10)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
sales_womens_dt = sales_womens_df.copy()
sales_womens_dt['invoice_date'] = pd.to_datetime(\
                        sales_womens_dt['invoice_date'], errors='coerce')

sales_women_dt_pv = sales_womens_dt.pivot_table(\
                                index=['invoice_date'], 
                                values='total_sales',
                                aggfunc = 'sum')
# sales_women_dt_pv.head()


sales_women_dt_pv= sales_women_dt_pv.rename(\
                columns={'total_sales':'Total Sales'})

# Sort the resampled pivot table in descending order on "Total Sales".
sales_women_dt_pv.sort_values('Total Sales', ascending=False).head(10)

```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
total_sales_weekly = sales_women_dt_pv.resample('W').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
total_sales_weekly.sort_values("Total Sales", ascending=False).head(10)
```
